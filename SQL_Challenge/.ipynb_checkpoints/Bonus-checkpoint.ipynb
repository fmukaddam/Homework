```python
#Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from config import username, password
```

```python
#Create Database Connecttion
from sqlalchemy import create_engine
engine = create_engine(f'postgresql://{username}:{password}@localhost:5432/employees_db')
connection = engine.connect()
```

```python
#Read the employee_db file
employee_data = pd.read_sql("SELECT * FROM employees", connection)
employee_data.head()
```

```python
salary_data = pd.read_sql("SELECT * FROM salaries", connection)
salary_data["salary"] = salary_data["salary"].str.replace("$", "")
salary_data["salary"] = salary_data["salary"].str.replace(",", "")
salary_data["salary"] =salary_data["salary"].astype("float")
salary_data.head()
```

```python
merged_df= employee_data.merge(salary_data, on="employee_no")
merged_df.head()
```

```python
titles_data = pd.read_sql("SELECT * FROM titles", connection)
titles_data
```

```python
final_merged= merged_df.merge(titles_data, on="emp_title_id")
final_merged.head()
```

```python
title_salary = final_merged[["employee_no", "salary"]]
title_salary.head()
```

```python
#Create a histogram
plt.figure(figsize= (12,6))
plt.hist(salary_data["salary"], color= "Blue", bins=20)
plt.ylabel("Number of Employees")
plt.xlabel("Salary($)")
plt.title("Frequency Distribution of Salary Ranges for Employees")
plt.grid()
plt.savefig("EmployeeSQL/Frequency Distribution of Salary Ranges for Employees.png")
plt.show()
```

```python
#Rename column emp_title_id
renamed_db = employee_data.rename(columns={"title_id": "emp_title_id"})
renamed_db.head()
```

```python
#Merge Title and Renamed Employee Table
employee_title = pd.merge(renamed_db, titles_data, on="emp_title_id", how="inner")
employee_title.head()
```

```python
#Average Salary
avg_salary = final_merged[["title", "salary"]].groupby(["title"]).mean().reset_index()
avg_salary                                                       
```

```python
fig1, ax1 = plt.subplots(figsize=(10,6))
x_axis = avg_salary["title"]
ticks= np.arange(len(x_axis))
y_axis= avg_salary["salary"]

#Create a chart based on the Data above
plt.bar(x_axis, y_axis, align="center", alpha= 0.8, color="Red")
plt.xticks(ticks,x_axis, rotation="vertical")
plt.ylabel("Salary($)")
plt.xlabel("Employee Title")
plt.title("Average Salary by Title")
plt.savefig("EmployeeSQL/average_salary.png")
plt.show()
```
